## DuckDB Sandbox
- An area for testing

https://duckdb.org/docs/current/clients/python/overview

```python
import duckdb
```

```python
# create a connection to a file called 'burger.db'
con = duckdb.connect("../burger.db") # Creates db automatically

# create a table and load data into it
con.sql("CREATE OR REPLACE TABLE test (i INTEGER)")
con.sql("INSERT INTO test VALUES (42)")

# query the table
con.table("test").show()

# explicitly close the connection
con.close()
# Note: connections also closed implicitly when they go out of scope
```
